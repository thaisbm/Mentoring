```python
import numpy as np 
import pandas as pd
```

```python
df_can = pd.read_excel('C:\Canada.xlsx',
                      sheet_name='Canada by Citizenship',
                      skiprows=range(20),
                      skipfooter=2)
print ('Data read into a pandas dataframe')
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)

df_can.head()
```

```python
#Rename
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.head()
```

```python
df_can.columns = list(map(str, df_can.columns))
#checking
all(isinstance(column, str) for column in df_can.columns)
```

```python
#set the country name as index - useful for quickly looking up countries using .loc method

df_can.set_index('Country', inplace=True)
df_can.head()
```

```python
#Add total column

df_can['Total'] = df_can.sum(axis=1)
df_can.head()
```

```python
#list of years from 1980 - 2013

years = list(map(str, range(1980, 2014)))

years
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')
```

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)

df_top5 = df_can.head()
df_top5 = df_top5[years].transpose()

df_top5.head()
```

```python
df_top5.index = df_top5.index.map(int) #change the index values to integer for plotting
df_top5.plot(kind='area',
            stacked=False,  #to produce an unstacked plot
            figsize=(20, 10),
            )

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
df_top5.index = df_top5.index.map(int) #change the index values to integer for plotting
df_top5.plot(kind='area',
             alpha=0.25, #0-1, default value a=0.5
            stacked=False,  #to produce an unstacked plot
            figsize=(20, 10),
            )

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
df_least5 = df_can.tail(5)
df_least5 = df_least5[years].transpose()
df_least5.head()

df_least5.index = df_least5.index.map(int)
df_least5.plot(kind='area', alpha=0.45, figsize=(20, 10))

plt.title('Immigration Trend of 5 Countries with least Contribution to Immigration')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
#Total number of immigrants to Canada from the top 15 countries - Horizontal Bar Plot

df_can.sort_values(by='Total', ascending=True, inplace=True)
df_top15 = df_can['Total'].tail(15)
df_top15
```

```python
df_top15.plot(kind='barh', figsize=(12, 12), color='steelblue')
plt.xlabel('Number of Immigrants')
plt.title('Top 15 Countries Contributing to the Immigration to Canada between 1980 - 2013')

#annotate value labels to each country
for index, value in enumerate(df_top15):
    label = format(int(value), ',')

plt.annotate(label, xy=(value - 47000, index - 0.10), color='white')

plt.show()
```
