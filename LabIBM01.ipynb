```python
# in pandas axis=0 represents rows (default) and axis=1 represents columns
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
df_can.head(2)
```

```python
#rename the columns
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.columns
```

```python
# Add a 'Total' column that sums up the total immigrants by country over the entire period 1980-2013
df_can['Total'] = df_can.sum(axis=1)
```

```python
#covenverting the years to string
df_can.columns = list(map(str, df_can.columns))
```

```python
#declare variable that will allow us to easily call upon the full range of years
years = list(map(str, range(1980, 2014)))
years
```

```python
#To filter the df based on condition, we simply pass the condition as a boolean vector. To show the data on Asian countries (AreaName = Asia)

condition = df_can['Continent'] == 'Asia'
print(condition)
```

```python
#pass this condition into the dataframe
df_can[condition]
```

```python
#we can pass multiple criteria in the same line.
#let´s filter for AreaName = Asia and RegName = Southern Asia

df_can[(df_can['Continent']=='Asia') & (df_can['Region']=='Southern Asia')]
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#Plot a line graph of immigration from Haiti using df.plot(). First we will extract the data series of Haiti

haiti = df_can.loc['Haiti', years]
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.index = haiti.index.map(int) #change the index values of Haiti to type integer for plotting
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

#annotate the 2010 Earthquake
#syntax: plt.text(x,y, label)
plt.text(2000, 6000, '2010 Earthquake')

plt.show()
```

```python
#Get the dataset for China and India

df_CI = df_can.loc[['India', 'China'], years]
df_CI.head()
```

```python
df_CI = df_CI.transpose() #transpose the df to swap the rows and columns
df_CI.head()
```

```python
df_CI.index = df_CI.index.map(int)
df_CI.plot(kind='line')

plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
#Compare the trend of top 5 countries that contributed the most to immigration to Canada

df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = df_can.head(5)
df_top5 = df_top5[years].transpose()
print(df_top5)

df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line', figsize=(14,8))
plt.title('Immigration trend of top 5 countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```
